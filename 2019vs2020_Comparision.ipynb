```python
#Import Depencies
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
#File of Comtrade
file = 'TMEC trade Ag.csv'
trade_data = pd.read_csv(file)
trade_df=pd.DataFrame(trade_data)
trade_df.replace(',','', regex=True, inplace=True)
trade_df.columns
```

```python
#Getting the totals for 2019 and 2020
Columns = ['2016-01', '2016-02', '2016-03', '2016-04', '2016-05', '2016-06',
       '2016-07', '2016-08', '2016-09', '2016-10', '2016-11', '2016-12',
       '2017-01', '2017-02', '2017-03', '2017-04', '2017-05', '2017-06',
       '2017-07', '2017-08', '2017-09', '2017-10', '2017-11', '2017-12',
       '2018-01', '2018-02', '2018-03', '2018-04', '2018-05', '2018-06',
       '2018-07', '2018-08', '2018-09', '2018-10', '2018-11', '2018-12',
       '2019-01', '2019-02', '2019-03', '2019-04', '2019-05', '2019-06',
       '2019-07', '2019-08', '2019-09', '2019-10', '2019-11', '2019-12',
       '2020-01', '2020-02', '2020-03', '2020-04', '2020-05', '2020-06',
       '2020-07', '2020-08', '2020-09', '2020-10', '2020-11', '2020-12']
trade_df[Columns] = trade_df[Columns].apply(pd.to_numeric, errors = 'coerce')
trade_df['Total 2019'] = trade_df.fillna(0)['2019-01'] + trade_df.fillna(0)['2019-02']+ trade_df.fillna(0)['2019-03']+ trade_df.fillna(0)['2019-04']\
                        + trade_df.fillna(0)['2019-05'] + trade_df.fillna(0)['2019-06']+ trade_df.fillna(0)['2019-07']+ trade_df.fillna(0)['2019-08']\
                        + trade_df.fillna(0)['2019-09']+ trade_df.fillna(0)['2019-10']+ trade_df.fillna(0)['2019-11']+ trade_df.fillna(0)['2019-12']
trade_df['Total 2020'] = trade_df.fillna(0)['2020-01'] + trade_df.fillna(0)['2020-02']+ trade_df.fillna(0)['2020-03']+ trade_df.fillna(0)['2020-04']\
                        + trade_df.fillna(0)['2020-05'] + trade_df.fillna(0)['2020-06']+ trade_df.fillna(0)['2020-07']+ trade_df.fillna(0)['2020-08']\
                        + trade_df.fillna(0)['2020-09']+ trade_df.fillna(0)['2020-10']+ trade_df.fillna(0)['2020-11']


trade_df
```

```python
#Extract just the inforation for totals analysis
clean_trade_df = trade_df[['Trade flow', 'Commodity', 'Partner', 'Reporter','Total 2019','Total 2020']]
clean_trade_df
```

```python
all_hs_df = clean_trade_df.loc[(clean_trade_df['Commodity'] == 'TOTAL Total of all HS commodities'),:]
world_df = all_hs_df.loc[(clean_trade_df['Partner'] == 'World'),:]
world_df_imports = world_df.loc[(world_df['Trade flow'] == 'Imports'),:]
world_df_exports= world_df.loc[(world_df['Trade flow'] == 'Exports'),:]
```

```python
#Extract the data for month vs month analysis of 2019 and 2020.
trade_df_2 = trade_df[['Trade flow', 'Commodity', 'Partner', 'Reporter',
       '2019-01', '2019-02', '2019-03', '2019-04', '2019-05', '2019-06',
       '2019-07', '2019-08', '2019-09', '2019-10','2020-01', '2020-02', '2020-03', '2020-04', '2020-05', '2020-06',
       '2020-07', '2020-08', '2020-09', '2020-10']]
trade_df_Comtotal = trade_df_2.loc[(trade_df_2['Partner'] == 'World') & (trade_df_2['Commodity'] == 'TOTAL Total of all HS commodities') ,:]
trade_df_Comtotal
```

```python
#Getting the data just for USA
trade_df_Comtotal_EU = trade_df_Comtotal.loc[(trade_df_Comtotal['Reporter']=='USA'),:]
trade_df_Comtotal_EU = trade_df_Comtotal_EU[['Trade flow','2019-01','2020-01','2019-02','2020-02', '2019-03','2020-03', 
                                             '2019-04', '2020-04','2019-05', '2020-05','2019-06','2020-06','2019-07','2020-07',
                                             '2019-08','2020-08','2019-09','2020-09','2019-10','2020-10'
                                             ]].set_index('Trade flow').T.rename_axis('Date').reset_index()

trade_df_Comtotal_EU["Imports"] = trade_df_Comtotal_EU["Imports"].astype(float)
trade_df_Comtotal_EU["Exports"] = trade_df_Comtotal_EU["Exports"].astype(float)
```

```python
#Plotting the data
sns.set_theme(style="whitegrid")
f, ax = plt.subplots(figsize=(15,10))

sns.set_color_codes("pastel")
sns.barplot(x='Date', y='Imports', data=trade_df_Comtotal_EU,
            label="Imports", color="b")
sns.set_color_codes("bright")
sns.barplot(x="Date", y='Exports', data=trade_df_Comtotal_EU,
            label="Exports", color="b")
ax.legend(ncol=2, loc="upper right", frameon=True)
ax.set( ylabel="",
       title="USA´s imports and exports 2019 vs 2020")
sns.despine(left=True, bottom=True)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.annotate('Start of Covid', xy=(5, 2.107566e+11), xytext=(4.5, 2.307566e+11),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('Pick of Covid Recovered', xy=(19, 2.107566e+11), xytext=(15, 2.000000e+11),
            arrowprops=dict(facecolor='black', shrink=0.05))

```
